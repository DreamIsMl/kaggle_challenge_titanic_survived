```python
from sklearn.model_selection import GridSearchCV
from sklearn.naive_bayes import GaussianNB
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
from sklearn.metrics import accuracy_score

# Assuming xtrain, ytrain, xtest, ytest are your training and testing data
# Assuming your categorical columns are at indices 1, 6, and 7

# Step 1: Define the ColumnTransformer
step1 = ColumnTransformer(transformers=[
    ('con_tnf', OneHotEncoder(sparse=False, drop='first', handle_unknown='ignore'), [1, 6, 7])
], remainder='passthrough')

# Step 2: Define the Gaussian Naive Bayes model
step2 = GaussianNB()

# Step 3: Create the Pipeline
pipe1 = Pipeline([
    ('step1', step1),
    ('step2', step2)
])

# Step 4: Define hyperparameters to tune
param_grid = {
    'step2__var_smoothing': [1e-9, 1e-8, 1e-7, 1e-6, 1e-5]
}

# Step 5: Create GridSearchCV
grid_search = GridSearchCV(pipe1, param_grid, cv=5, scoring='accuracy', verbose=1)

# Step 6: Fit the model with hyperparameter tuning
grid_search.fit(xtrain, ytrain)

# Step 7: Get the best estimator from the search
best_pipe = grid_search.best_estimator_

# Step 8: Predict and evaluate
y_pred = best_pipe.predict(xtest)
accuracy = accuracy_score(ytest, y_pred)

# Step 9: Print results
print('Best Accuracy:', accuracy)
print('Best Hyperparameters:', grid_search.best_params_)
```

```python
Xtrain,Xtest,Ytrain,Ytest = train_test_split(x,y,test_size=.2,random_state=305)
from sklearn.model_selection import GridSearchCV
from sklearn.naive_bayes import GaussianNB
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
from sklearn.metrics import accuracy_score

# Assuming xtrain, ytrain, xtest, ytest are your training and testing data
# Assuming your categorical columns are at indices 1, 6, and 7

# Step 1: Define the ColumnTransformer
step1 = ColumnTransformer(transformers=[
    ('con_tnf', OneHotEncoder(sparse=False, drop='first', handle_unknown='ignore'), [1, 6, 7])
], remainder='passthrough')

# Step 2: Define the Gaussian Naive Bayes model
step2 = GaussianNB(var_smoothing=1e-07)  # Use the var_smoothing parameter here

# Step 3: Create the Pipeline
final_pipe = Pipeline([
    ('step1', step1),
    ('step2', step2)
])

# Step 4: Fit the model
final_pipe.fit(Xtrain, Ytrain)

# Step 5: Predict and evaluate
y_pred = final_pipe.predict(Xtest)
accuracy = accuracy_score(Ytest, y_pred)

# Step 6: Print results
print('Accuracy:', accuracy)
print("precision score", precision_score(Ytest,y_pred))
```
